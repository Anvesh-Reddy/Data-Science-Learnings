```python
from tpot import TPOTClassifier, TPOTRegressor
from sklearn.datasets import load_digits, load_boston
from sklearn.model_selection import train_test_split
```

### Classification

```python
digits = load_digits()
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target)
```

```python
tpot_obj_clf = TPOTClassifier(generations=5, population_size=20, verbosity=2)
tpot_obj_clf.fit(X_train, y_train)
```

```python
tpot_obj_clf.score(X_test, y_test)
```

```python
tpot_obj_clf.fitted_pipeline_
```

```python
tpot_obj_clf.evaluated_individuals_
```

```python
tpot_obj_clf.export('tpot_digits_pipeline.py')
```

### Regression

```python
boston = load_boston()
X_train_boston, X_test_boston, y_train_boston, y_test_boston = train_test_split(boston.data, boston.target)
```

```python
tpot_obj_rgr = TPOTRegressor(template='Selector-Transformer-Regressor', generations=5, population_size=20, verbosity=2)
tpot_obj_rgr.fit(X_train_boston, y_train_boston)
```

```python
tpot_obj_rgr_specific = TPOTRegressor(template='StandardScaler-OneHotEncoder-Regressor', generations=5, population_size=20, verbosity=2)
tpot_obj_rgr_specific.fit(X_train_boston, y_train_boston)
```

```python
tpot_obj_rgr.score(X_test_boston, y_test_boston)
```

```python
tpot_obj_rgr.fitted_pipeline_
```

```python
tpot_obj_rgr.evaluated_individuals_
```

### TPOT using Config file

```python
import json
```

```python
with open('tpot_digits_config.py', 'r') as f:
    tpot_config = json.loads(f.read())
```

```python
tpot = TPOTClassifier(generations=5, population_size=20, verbosity=2,
                      config_dict=tpot_config)
tpot.fit(X_train, y_train)
```

```python
tpot.score(X_test, y_test)
```
